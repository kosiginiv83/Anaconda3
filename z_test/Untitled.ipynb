```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import datetime
import math
import operator

from pprint import pprint

%matplotlib inline
```

```python
date = []
value = []
with open('Benchmark.txt', encoding='utf-8') as f:
    for line in f:
        line_list = line.split(';')
        date.append(line_list[0])
        value.append(line_list[1].strip())
        
date.remove('Date')
value.remove('Our_L')

value_int = list(map(int, value))
date_time = list(map(lambda x: time.strptime(date[x], '%d-%m-%y %H:%M'), range(len(date))))

benchmark_df = pd.DataFrame({
    'date': date_time,
    'value': value
})

print(benchmark_df[8:11], '\n')
print(benchmark_df[8:11]['date'], '\n')
print(benchmark_df.iloc[0]['date'], '\n')
print(time.mktime(benchmark_df.iloc[0]['date']), '\n')
```

```python
dpi = 80
fig = plt.figure(dpi = dpi, figsize = (1240 / dpi, 512 / dpi))
date_s = list(map(lambda x: time.mktime(benchmark_df.iloc[x]['date']), range(len(benchmark_df))))
plt.plot(date_s, benchmark_df['value'])
plt.show()
```
